```python
def preprocess_text(text):
    text = text.lower()

    sentences = sent_tokenize(text)
    processed_sentences = []
    for sentence in sentences:
        words = word_tokenize(sentence)
        words = [word for word in words if word not in stop_words and word not in string.punctuation]
        words = [stemmer.stem(word) for word in words]
        processed_sentences.append(' '.join(words))
    
    return processed_sentences
```

```python
from collections import Counter
import nltk

# Normalization step for term frequencies
def compute_normalized_tf(sentences):
    """Compute and normalize term frequency (TF) scores."""
    word_count = Counter()
    total_word_count = 0
    for sentence in sentences:
        words = sentence.split()
        word_count.update(words)
        total_word_count += len(words)
    # Normalize by total word count
    return {word: count / total_word_count for word, count in word_count.items()}

def score_sentences(preprocessed_sentences, tf_scores):
    """Score sentences based on normalized TF scores."""
    sentence_scores = {}
    for idx, sentence in enumerate(preprocessed_sentences):
        words = sentence.split()
        # Sum of normalized word scores
        score = sum(tf_scores.get(word, 0) for word in words)
        sentence_scores[idx] = score
    return sentence_scores

def get_summary(original_sentences, preprocessed_sentences, sentence_scores, max_words=100):
    """Generate summary from top-ranked sentences."""
    # Rank sentences based on their scores
    ranked_sentences = sorted(sentence_scores.items(), key=lambda x: x[1], reverse=True)
    
    summary = []
    word_count = 0
    
    for idx, _ in ranked_sentences:
        sentence_word_count = len(original_sentences[idx].split())
        if word_count + sentence_word_count <= max_words:
            summary.append(original_sentences[idx])  # Use original sentences
            word_count += sentence_word_count
        else:
            break
    
    return ' '.join(summary)

extractive_summaries = []

# Process each document
for original_doc, preprocessed_doc in zip(docs, preprocessed_docs):
    original_sentences = nltk.sent_tokenize(original_doc)
    
    # Normalize term frequencies
    tf_scores = compute_normalized_tf(preprocessed_doc)

    # Score sentences using normalized TF
    sentence_scores = score_sentences(preprocessed_doc, tf_scores)
    
    # Generate concise summary with a limit of 100 words
    summary = get_summary(original_sentences, preprocessed_doc, sentence_scores, max_words=100)
    extractive_summaries.append(summary)

print("Extractive summaries with a 100-word limit generated for all documents.")
```

```python
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
from collections import Counter
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import string

stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def preprocess_text(text):
    text = text.lower()
    words = word_tokenize(text)
    words = [word for word in words if word not in stop_words and word not in string.punctuation]
    words = [stemmer.stem(word) for word in words]
    return words

def compute_tf(words):
    word_count = Counter(words)
    total_words = sum(word_count.values())
    return {word: count / total_words for word, count in word_count.items()}

def compute_kld(doc_dist, sent_dist):
    kld = 0
    for word, prob in doc_dist.items():
        if word in sent_dist and sent_dist[word] > 0:
            kld += prob * np.log(prob / sent_dist[word])
        else:
            kld += prob * np.log(prob / 1e-10) 
    return kld

def kld_greedy_summarization(doc_text, max_words=100):

    sentences = sent_tokenize(doc_text)
    processed_sentences = [preprocess_text(sentence) for sentence in sentences]

    doc_words = preprocess_text(doc_text)
    doc_tf = compute_tf(doc_words)

    sentence_tfs = [compute_tf(sentence) for sentence in processed_sentences]

    selected_sentences = []
    selected_indices = []
    combined_tf = Counter()
    word_count = 0

    while word_count < max_words and len(selected_indices) < len(sentences):
        best_kld = float('inf')
        best_sentence_idx = None
        
        for i, sent_tf in enumerate(sentence_tfs):
            if i in selected_indices:
                continue  

            combined_sentence_tf = combined_tf.copy()
            for word, count in sent_tf.items():
                combined_sentence_tf[word] += count

            total_count = sum(combined_sentence_tf.values())
            combined_sentence_dist = {word: count / total_count for word, count in combined_sentence_tf.items()}

            kld = compute_kld(doc_tf, combined_sentence_dist)
            
            if kld < best_kld:
                best_kld = kld
                best_sentence_idx = i

        if best_sentence_idx is not None:
            selected_indices.append(best_sentence_idx)
            selected_sentences.append(sentences[best_sentence_idx])
            word_count += len(processed_sentences[best_sentence_idx])
            

            for word, count in sentence_tfs[best_sentence_idx].items():
                combined_tf[word] += count
    
    return ' '.join(selected_sentences)

docs = []
human_summaries = []

for filename in sorted(os.listdir(human_path)):
    file_path = os.path.join(human_path, filename)
    if filename.endswith(".txt"):
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            if "human_summary" in filename:
                human_summaries.append(content)
            else:
                docs.append(content)

for filename in sorted(os.listdir(original_path)):
    file_path = os.path.join(original_path, filename)
    if filename.endswith(".txt"):
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            if "human_summary" in filename:
                human_summaries.append(content)
            else:
                docs.append(content)


df = pd.DataFrame({
    'document': docs,
    'human_summary': human_summaries
})


summaries = []
for doc in tqdm(df['document']):
    summary = kld_greedy_summarization(doc, max_words=100)
    summaries.append(summary)


df['extractive_summary'] = summaries

```
